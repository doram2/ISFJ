```python
bank = []
def check(bank, account):
    value = False
    for i in range(len(bank)):
        if(bank[i][0] == account):
            value = True
            break;
    return value
    
while True:
    print()
    print("=="*20)
    print("계좌등록: 1 | 입금: 2 | 출금: 3 | 계좌출력: 4 | 종료 : 5")
    num = int(input("해당번호를 입력해주세요. : "))
    print("=="*20)
    print()
    if num == 1:
        print("계좌등록 페이지입니다.")
        while True:
            account = int(input("계좌번호를 입력해주세요. : "))
            if(check(bank,account)):
                print("중복된 계좌번호입니다. 새로운 계좌를 등록해주세요.")
                continue
            else:
                break;
        money = int(input("입금 금액을 입력해주세요. : "))
        name = input("계좌주를 입력해주세요. : ")
        bank.append([account, money, name])
   
    if num == 2:
        print("입금페이지입니다.")
        while True:
            deposit = int(input("입금 계좌를 입력해주세요. : "))
            if(check(bank,deposit) == False):
                print("존재하지 않는 계좌번호입니다.")
                continue
            else:
                break
        money = int(input("입금 금액을 입력해주세요. : "))
        for i in range(len(bank)):
            if(bank[i][0] == deposit):
                bank[i][1] += money
                break
    
    if num == 3:
        print("출금페이지입니다.")
        while True:
            withdrawal = int(input("출금 계좌를 입력해주세요. : "))
            if(check(bank,withdrawal) == False):
                print("존재하지 않는 계좌번호입니다.")
                continue
            else:
                break
        money = int(input("출금 금액을 입력해주세요. : "))
        for i in range(len(bank)):
            if(bank[i][0] == deposit and bank[i][1] >= money):
                bank[i][1] -= money
                break
            else:
                print("출금 가능한 금액이 부족합니다.")
    
    if num == 4:
        if(len(bank) != 0):
            print("계좌 출력페이지입니다.\n")
            print("\t계좌번호\t 계좌주\t 금액\t")
            print("--"*20)
            for i in range(len(bank)):
                print("\t",bank[i][0],"\t",bank[i][2],"\t",bank[i][1])
        else:
            print("출력할 계좌가 없습니다. 계좌를 등록해주세요.")
    
    if num == 5:
        print("프로그램을 종료합니다.")
        break;
```
